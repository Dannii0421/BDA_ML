파생변수 만들기

- 현재의 데이터들을 이용해서 분석도 가능하지만 -> 변수들을 조합하여서 새로운 변수를 만들어 분석할 수 있다.
- 영어, 수학, 과학 점수가 있으면 -> 이 세 개의 변수를 평균으로 바꿔서 하나의 파생변수로도 나타낼 수 있다.
- BDA 상점, 벌점 상점 - 벌점 = 수료가능한 점수
      - 판다스는 직관적으로 만들 수 있다.
      - 판다스의 함수를 이용해서도 만들 수 있다.

```python
df=pd.DataFrame({'이름':['홍길동','박길동','오길동','이길동'],
                '상점': [10,20,30,40],
                 '벌점': [30,5,15,20]})
```

```python
df_1 = df.copy()
```

```python
df_1['수료가능점수'] = df_1['상점'] - df_1['벌점']
## 상점 - 벌점으로 만들면 되는 거 아닌가?
```

```python
##수료, 수료불가
##수료가능점수 기준에 따라 수료, 수료불가
for i in df_1['수료가능점수']:
    if i > 0:
        print('수료')
    else:
        print('수료불가')
```

```python
## numpy를 가지고 인덱스를 쉽게 찾아서 필요한 전처리를 진행할 수 있다.
## np.where
## 참이 앞의 값으로 매핑, 거짓이 뒤 값으로 매핑

df_1['수료여부'] = np.where(df_1['수료가능점수']>0,'수료','수료불가')
```

```python
df_1
```

```python
df_1['수료가능점수']>0
```

```python
## 범주형 데이터를 카운팅 하는 간단한 방법
## value_counts()
df_1['수료여부'].value_counts()
```

loc[], iloc[]

- 인덱싱 작업
- loc:Location(인덱스)-처음 생성된 데이터프레임의 인덱스
- iloc:Integer Location(넘버링된 인덱스)-현재 기준의 데이터프레임의 순서
  인덱스(데이터프레임을 조정하거나 재조립하게 되면 그 기준 순서로 넘버링)
   - 데이터프레임.loc[인덱스]
   - 데이터프레임.iloc[인덱스]
- loc, iloc 은 행과 열 모두 접근 가능하다.
- loc : 컬럼명을 명시적으로 적어줘야 한다.
- iloc : 뒤에도 컬럼을 모두 다 숫자로 int로 적어줘야 한다. 안 그러면 에러가 난다

```python
## 슬라이싱 range() 사용
## loc[행,열]
## loc[행:행,열:열]
tt.loc[10:30:2,['survived','pclass']]
```

```python
tt.iloc[10:30:2,[0,1]]
```

```python
tt
```

```python
tt.iloc[10:30:2,3:9]
```

방금 위에서 배운 loc,iloc 그리고 처음 배웠던 [] [[]] 컬럼 추출을 응용하자

```python
## 먼저 열을 선택한 후, 행을 선택
tt[['survived','pclass','sex','fare']].loc[4] 
```

```python
## 먼저 행을 선택한 후 열을 선택
tt.loc[4][['survived','pclass','sex','fare']]
```

데이터 분석에 필요한 기초 문법

- query() 데이터 조건에 따른 추출
- df[df[]] 불린 문법
- sort_values() 정렬
- 파생변수 추가 (assign,lambda) 그 파생변수를 함수를 써서 만들기
- groupby - 요약 집계 -> agg함수 같이 데이터를 집계해서 사용한다.
- merge - 데이터 병합
- str - 문자열 문법
- 기타(시계열 등등)특정한 데이터 타입에 따른 문법

```python
Query

- 쿼리 내가 원하는 데이터를 추출하기 위해 조건을 질의하는 것
- 그 조건을 명확하게 문법으로 풀어서 적고 추출하는 것
- 데이터프레임.query('조건')
```

```python
## pclass가 3인 승객을 추출해줘
## 그런데 여기에 fare 5 이상인 승객을 출력해줘
## 숫자로만 데이터를 추출
tt.query('pclass==3 & fare >= 5')
```

```python
## 문자로 데이터 추출
## sex male인데 여기에 age는 25살 이상으로 추출해줘
tt.query("sex == 'male' & age >= 25")
```

```python
tt.query("sex =='male' or age>=25")
```
