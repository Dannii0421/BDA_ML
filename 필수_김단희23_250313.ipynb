```python
import pandas as pd
import seaborn as sns

tt=sns.load_dataset('titanic')
```

```python
## 1. 나이가 30세 이상인 승객만 추출하기
tt.query('age >= 30')
```

```python
## 2. 성별이 여성(female)이고, 생존한(Survived == 1) 승객 추출하기
tt.query("sex == 'female' &  survived == 1")
```

```python
## 3. 1등석에 탑승한 승객의 이름과 운임(Fare)만 출력하기
tt.query("pclass == 1")[['fare']]
```

```python
## 4. 나이가 10세 이하인 승객의 이름과 나이만 출력하기
tt.query("age <= 10")[['sex','age']]
```

```python
## 5. 2등석 승객 중 요금이 20달러 이상 낸 승객 추출하기
tt.query("pclass == 2 & fare >= 20")
```

```python
## 6. 운임(Fare)이 가장 높은 상위 10명의 데이터 출력하기
tt.sort_values(by='fare', ascending=False).head(10)
```

```python
## 7. 생존한 남성 중에서 나이가 가장 어린 승객 정보 출력하기
tt.query("survived == 1 & sex == 'male'").sort_values(by='age').head(1)
```
